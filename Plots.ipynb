The Bifurcation diagram code is written in fortran90 by the following algorithm

```fortran
program Chaos   
    implicit none   
    integer :: n   
    real :: x, r   

    ! Set initial condition
    x = 0.4

    ! Loop over r values
    r = 0.7
    do while (r <= 1)
        do n = 1, 1100
            x = 4.0 * r * x * (1.0 - x)
            if (n > 1000) then  ! Only print after reaching a certain iteration
                print *, r, x
            endif
        end do
        r = r + 0.001
    end do

end program Chaos


As can been seen the key component is to let the values converge by performing a $1000$ steps first and then because the behavior will have scattered values for one value of $r$, $100$ iteration are recorded to produce a nice plot.  

#### Bifurcation Point
We found that the value of $r$ at which the single fixed point $x$ bifurcates into two values 
$x_1$ and $x_2$ is  $r = \frac{3}{4} = 0.75 $. This pair of $x$ values, $x_1$ and 
$x_2$, forms a stable attractor of period $2$.

These findings demonstrate the complex and fascinating dynamical properties of the logistic map and 
highlight the presence of stable cycles of different periods as $r$ varies.

We also observe these stable attractors at $0.863$, and at values a little less than $0.890$. Periods go as $2$, $4$, $8$, as the existing branches keep splitting into two.
